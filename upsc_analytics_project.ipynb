```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10,6)
```

```python
df = pd.read_csv('upsc_question.csv')

print("Rows and Columns:", df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
df.isnull().sum()
```

Analysis 1: Subject Distribution

```python
subject_counts = df['Subject'].value_counts()

print(subject_counts)
```

```python
subject_counts.plot(kind='bar')

plt.title("UPSC Subject Distribution (2023-2026)")
plt.xlabel("Subject")
plt.ylabel("Number of Questions")
plt.xticks(rotation=45)
plt.show()
```

Analysis 2: Year-wise Subject Trend

```python
subject_year = pd.crosstab(
    df['Year'],
    df['Subject']
)

subject_year
```

```python
subject_year.plot(kind='bar')

plt.title("Year-wise Subject Trend")
plt.ylabel("Questions")
plt.show()
```

Analysis 3: Difficulty Analysis

```python
difficulty_counts = df['Difficulty'].value_counts()

difficulty_counts

```

```python
difficulty_counts.plot(kind='bar')

plt.title("Difficulty Distribution")
plt.ylabel("Questions")
plt.show()
```

Analysis 4: Difficulty Trend by Year

```python
difficulty_year = pd.crosstab(
    df['Year'],
    df['Difficulty']
)

difficulty_year
```

```python
difficulty_year.plot(kind='bar')

plt.title("Difficulty Trend by Year")
plt.ylabel("Questions")
plt.show()
```

Analysis 5: Static vs Current Affairs


```python
static_current = pd.crosstab(
    df['Year'],
    df['Static_Current']
)

static_current
```

```python
static_current.plot(kind='bar')

plt.title("Static vs Current Trend")
plt.ylabel("Questions")
plt.show()
```

Analysis 6: Question Type Evolution

```python
question_type = pd.crosstab(
    df['Year'],
    df['Question_Type']
)

question_type
```

```python
question_type.plot(kind='bar')

plt.title("Question Type Evolution")
plt.ylabel("Questions")
plt.show()
```

Analysis 7: Top 20 Most Asked Topics

```python
top_topics = df['Topic'].value_counts().head(20)

top_topics
```

```python
top_topics.plot(kind='bar')

plt.title("Top 20 Most Frequent Topics")
plt.ylabel("Frequency")
plt.show()
```

```python
upsc_2026 = df[df['Year'] == 2026]

upsc_2026['Subject'].value_counts()
```

```python
upsc_2026['Difficulty'].value_counts()
```
